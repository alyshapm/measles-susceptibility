```python
import pandas as pd
from sklearn import decomposition
```

```python
# file_names = ['matrix_1_1.csv', 'matrix_1_2.csv', 'matrix_7_1.csv', 'matrix_7_2.csv']
```

```python
df1 = pd.read_csv('https://raw.githubusercontent.com/alyshapm/measles-susceptibility/main/matrix/matrix_1_1.csv')
df2 = pd.read_csv('https://raw.githubusercontent.com/alyshapm/measles-susceptibility/main/matrix/matrix_1_2.csv')
df3 = pd.read_csv('https://raw.githubusercontent.com/alyshapm/measles-susceptibility/main/matrix/matrix_7_1.csv')
df4 = pd.read_csv('https://raw.githubusercontent.com/alyshapm/measles-susceptibility/main/matrix/matrix_7_2.csv')

all_df = [df1, df2, df3, df4]

for df in all_df:
  df.rename(columns={'Unnamed: 0': 'Sample'}, inplace=True)
```

```python
from functools import reduce
```

```python
merged_df = reduce(lambda left, right: pd.merge(left, right, on=['Sample', 'Population code'], how='inner'), all_df)
```

```python
merged_df
```

```python
# to check if the population code column still exists
merged_df['Population code'].nunique()
```

```python
merged_df.to_csv('merged_matrix.csv', index=False)
```

```python
non_snp_columns = ['Population code', 'Sample']
```

```python
df_snps = merged_df.drop(non_snp_columns, axis=1)
matrix = df_snps.to_numpy()
# print(matrix.shape)
matrix
```

```python
pca = decomposition.PCA(n_components=2)
pca.fit(matrix)
```

```python
print(pca.explained_variance_ratio_)
print(pca.singular_values_)
```

 The explained variance ratio tells you how much of the total variance in the genetic data is captured by each principal component.

 The singular values values can be thought of as indicating how strongly each principal component represents the patterns in the genetic data.

```python
to_plot = pca.transform(matrix)
to_plot.shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x=to_plot[:, 0], y=to_plot[:, 1])
```

```python
import altair as alt
```

```python
df_plot = merged_df[non_snp_columns].copy()
df_plot
```

```python
df_plot['PC1'] = to_plot[:, 0]
df_plot['PC2'] = to_plot[:, 1]
df_plot
```

Different colors appear to represent different superpopulations or ancestries. This can show if genetic variation related to measles susceptibility is more common in some ancestries than others.

```python
alt.Chart(df_plot).mark_point().encode(
    x='PC1',
    y='PC2',
    color=alt.Color('Population code', scale=alt.Scale(scheme='category20'))
)
```

```python
pop = pd.read_csv('https://raw.githubusercontent.com/alyshapm/measles-susceptibility/main/data/igsr_populations.tsv', sep='\t')
pop
```

```python
df_plot = df_plot.merge(pop, on='Population code', how='inner')
df_plot.head()
```

```python
alt.Chart(df_plot).mark_point().encode(
    x='PC1',
    y='PC2',
    color=alt.Color('Superpopulation name', scale=alt.Scale(scheme='category20')),
    fill='Population code'
)
```

The x-axis (PC1) and y-axis (PC2) represent the most significant genetic variations in the data. These axes might be capturing the combination of SNPs that explain the most variability in measles susceptibility among the individuals.


# tSNE

```python
from sklearn.manifold import TSNE
```

```python
X = matrix
X_embedded = TSNE(n_components=2, learning_rate='auto',
                  init='random').fit_transform(X)
X_embedded.shape
```

```python
df_plot['tsne1'] = X_embedded[:,0]
df_plot['tsne2'] = X_embedded[:,1]
```

```python
df_plot
```

### tSNE:

```python
alt.Chart(df_plot).mark_point().encode(
    x='tsne1',
    y='tsne2',
    color=alt.Color('Superpopulation name', scale=alt.Scale(scheme='category20'))
)
```

```python
alt.Chart(df_plot).mark_point().encode(
    x='tsne1',
    y='tsne2',
    color=alt.Color('Population code', scale=alt.Scale(scheme='category20'))
)
```

### PCA:

```python
alt.Chart(df_plot).mark_point().encode(
    x='PC1',
    y='PC2',
    color=alt.Color('Superpopulation name', scale=alt.Scale(scheme='category20'))
)
```

# Plotting geographic data

```python
df_plot.info()
```

```python
import numpy as np
import plotly.express as px
```

```python
def add_noise(value, noise_factor=8):
    return value + np.random.uniform(-noise_factor, noise_factor)
```

```python
df_noise = df_plot
df_noise['Population latitude'] = df_plot['Population latitude'].apply(add_noise)
df_noise['Population longitude'] = df_plot['Population longitude'].apply(add_noise)
```

```python
fig = px.scatter_geo(df_noise,
                     lat='Population latitude',
                     lon='Population longitude',
                     color='Superpopulation name',
                     title='Population Distribution of Measles-related SNPS Globally',
                     )

fig.show()
```
